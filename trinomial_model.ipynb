```python

import random
import math
import numpy as np
import scipy as sp
import pandas as pnd

```

Asumptions

```python
exp = math.exp
sqrt = math.sqrt
sigma = 0.25
```

europian call option


```python
def TrinomialTree (OptionType, CallType, S, K, T, r, sigma, n):
    optionvalue = np.repeat(0, 2*n+1)
    # print(optionvalue)
    opt = 1 if OptionType == "a" else 0
    # call-put type
    if (CallType == "c"):
        z = +1
    if (CallType == "p"):
        z = -1

    # time interval
    dt = T/n
    # up-down jump size
    u = exp(sigma * sqrt(2*dt))
    d = 1/u
    m = 1
    # R = exp ** (r*dt)

    #probabilities of going up and down
    # pd = ((exp(sigma*sqrt(dt/2))-exp(r*dt/2))/(exp(sigma*sqrt(dt/2)) - exp(-sigma*sqrt(dt/2))))**2
    # pu = ((exp(r*dt/2)-exp(-sigma*sqrt(dt/2)))/(exp(sigma*sqrt(dt/2)) - exp(-sigma*sqrt(dt/2))))**2
    # pm = 1-pd-pu
    a = math.exp((r)*dt/2)
    a1 = math.exp(sigma * math.sqrt(dt/2))
    a2 = math.exp(-sigma * math.sqrt(dt/2))
    pu = ((a - a2) / (a1 - a2))**2
    pd = ((a1 - a) / (a1 - a2))**2
    pm=1-pu-pd
    df = exp(-r*dt)
    R = math.exp(r * dt)
    da = []
    vals = [0 for i in range(2*n +1)]
    vals[n] = S
    # print(pu)
    da.append(vals.copy())
    for i in range(1,n+1):
        a = vals.copy()
        a[n-i] = a[n-i+1] * u
        a[n+i] = a[n+i-1] * d
        da.append(a.copy()) 
        vals = a
    da_cal = {}
    for i in range(da[n].__len__()):
        da_cal[(n,i)] = max((da[n][i] - S ) * z,0)
    for k in range(n-1,-1,-1):
        for i in range(0,2*k+1):
            a = (pd * da_cal[(k+1,i+2)] + pm * da_cal[(k+1,i+1)] + pu * da_cal[(k+1,i)]) / R
            # da_cal[(k,i)] = max(a ,max(da[k][n-k+i]-S,0) * opt)
            if(opt):
                da_cal[(k,i)] = max(a ,max((da[k][n-k+i]-S) * z,0) )
            else : 
                da_cal[(k,i)] = a

    # print(da_cal)
    dic = {}
    for i in range(len(da)):
        dic[i] = da[i]
    df = pnd.DataFrame(data=dic)
    df.to_csv('values.csv')
    # print(df)
    TrinomialTree = optionvalue[1]

    return  (da_cal[(0,0)],TrinomialTree, u, d, pu, pd, pm)
TrinomialTree(OptionType="e" ,CallType= "p", S = 100, K = 100, T = 3, r = 0.03, sigma = 0.2, n =9)
```
